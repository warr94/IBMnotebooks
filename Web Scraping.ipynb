```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

# Especificar url de la página web
source = urlopen('https://en.wikipedia.org/wiki/John_D._Hunter').read()

# Almacenar información mediante BeautifulSoup
soup = BeautifulSoup(source,'lxml')
soup
```

```python
# Extraiga el contenido de texto sin formato de los párrafos
paras = []
for paragraph in soup.find_all('p'):
    paras.append(str(paragraph.text))
    
print(paras)
```

```python
# Extraer texto de encabezados de párrafo
heads = []
for head in soup.find_all('span', attrs={'mw-headline'}):
    heads.append(str(head.text))
    
print(heads)
```

```python
# Intercalar párrafos y encabezados
text = [val for pair in zip(paras, heads) for val in pair]
text = ' '.join(text)

print(text)
```

```python
# Coloque los superíndices de las notas al pie entre paréntesis
text = re.sub(r"\[.*?\]+", '', text)

print(text)
```

```python
# Reemplace '\n' (una nueva línea) con '' y finalice la cadena en $1000.
text = text.replace('\n', '')[:-11]

print(text)
```
